```python
senti_list = []
def classify(data):
    for polarity in data:
        if polarity == 0:
            senti_list.append('neutral')
        elif polarity > 0:
            senti_list.append('positive')
        else:
            senti_list.append('negative')
```

# Data preprocessing

```python
import string
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

```python
def preprocess(text):
    headlines = [headline for headline in text.lower() if headline not in string.punctuation ]
    new_headlines = ''.join(headlines)
    new_headlines = [new_headline for new_headline in new_headlines.split() if 
                    new_headline not in stop_words]
    return ' '.join(new_headlines)
```

```python
def lemmatization(data):
    lemmatizer = WordNetLemmatizer()
    lemma = [lemmatizer.lemmatize(word) for word in data.split()]
    return ' '.join(lemma)
        
```

```python
df_all[['new_headline','lemmatized_headline']].iloc[20:25]
```

```python
def Stemming(data):
    steming = PorterStemmer()
    stem = [steming.stem(word) for word in data.split()]
    return ' '.join(stem)
```

```python
df_all['stemmed_headline'] = df_all['new_headline'].apply(Stemming)
```

```python
df_all['lemmatized_headline'] = df_all['new_headline'].apply(lemmatization)
```

```python
month = ['september','sep','oct','october','dec','december','jan','january','feb','february']
```

```python
def get_month(data):
    the_month = [mnt for mnt in data.split() if mnt.lower() in month]
    the_month = ' '.join(the_month)
    return the_month
```

```python
df_all_month = pd.DataFrame(df_all['date'].apply(get_month))
```

```python
df_all_month.head()
```

```python
df_all_month.tail()
```

```python
month_classes = []
def classify_month(data):
    for mnt in data:
        if (mnt.lower() == 'sep') or (mnt.lower() =='september'):
            month_classes.append('september')
        elif (mnt.lower() == 'oct') or (mnt.lower() == 'october'):
             month_classes.append('october')
        elif (mnt.lower() == 'nov') or (mnt.lower() =='november'):
            month_classes.append('november')
        elif (mnt.lower() == 'dec') or (mnt.lower() =='december'):
            month_classes.append('december')
        elif (mnt.lower() == 'jan') or (mnt.lower() == 'january'):
            month_classes.append('january')
        else:
            month_classes.append('february')
        
```

```python
classify_month(df_all_month['date'])
```

# splitting training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_all['lemmatized_headline']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
print(X_train.shape)
print(y_train.shape)
```

# creating supervised learning model

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc_pipe = Pipeline([
    ('cv',CountVectorizer()),
    ('tfidf',TfidfTransformer()),
    ('rfc', RandomForestClassifier())
])
```

```python
rfc_pipe.fit(X_train,y_train)
```

```python
rfc_prediction = rfc_pipe.predict(X_test)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc_pipe = Pipeline([
    ('cv',CountVectorizer()),
    ('tfidf',TfidfTransformer()),
    ('dtc',DecisionTreeClassifier())
])
```

```python
dtc_pipe.fit(X_train,y_train)
```

```python
dtc_prediction = dtc_pipe.predict(X_test)
```

```python
rfc_result = pd.DataFrame({'Actual value':y_test,'predited value':rfc_prediction})
rfc_result.head()
```

```python
dtc_result = pd.DataFrame({'Actual value':y_test,'predited value':dtc_prediction})
dtc_result.head()
```
